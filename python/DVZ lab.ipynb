# 1a

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path = r"company-sales.csv"
data = pd.read_csv(path)

plt.figure(figsize = (10, 6))
plt.plot(data.month_number, data.total_profit, color = "green", linestyle = "dotted", marker = 's')
plt.xlabel("Month")
plt.ylabel("Total profit")
plt.title("Total profit overtime")
for i in range(len(data)):
    plt.text(data.month_number[i], data.total_profit[i], 
             f"{data.total_profit[i]:,.2f}",
             ha="left", size=8)
plt.xticks()
plt.yticks()
plt.grid(True)
```

# 3a

```python
# Calculate total sales for each product (excluding 'month_number' and 'total_profit')
product_columns = [col for col in data.columns if col not in ['month_number', 'total_profit']]
total_sales = data[product_columns].sum()

# Find the index of the highest sale for explode
explode = [0] * len(total_sales)
max_index = total_sales.argmax()
explode[max_index] = 0.1

# Create pie chart
plt.figure(figsize=(8, 6))
patches, texts, autotexts = plt.pie(
    total_sales,
    labels=product_columns,
    autopct=lambda pct: f"{int(round(pct/100.*total_sales.sum()))}",
    explode=explode,
    startangle=60
)
# Set font size for labels and values
for text in texts + autotexts:
    text.set_fontsize(10)
plt.title("Total Sales for Each Product (Last Year)")
plt.show()
```

# 3b

```python
# Read only face cream and facewash sales data
face_cream_facewash = data[['month_number', 'facecream', 'facewash']]

# Plot horizontal bar chart for both products
plt.figure(figsize=(10, 6))
plt.barh(face_cream_facewash['month_number'] - 0.2, face_cream_facewash['facecream'], height=0.4, label='Face Cream', color='skyblue')
plt.barh(face_cream_facewash['month_number'] + 0.2, face_cream_facewash['facewash'], height=0.4, label='Face Wash', color='salmon')
plt.xlabel('Sales units per month')
plt.ylabel('Month Number')
plt.title('Monthly Sales of Face Cream and Face Wash')
plt.yticks(face_cream_facewash['month_number'])
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

languages = ['Fict', 'Tech', 'Moti', 'Business', 'Nutri', 'Dev']
sales = [5.2, 19.6, 8, 8.7, 7.3, 3.7]

colors = plt.cm.viridis(np.linspace(0, 1, len(languages)))
bars = plt.barh(languages, sales, color=colors)
plt.xticks(range(0, 21, 2))
plt.yticks(languages)
for bar, lang in zip(bars, languages):
    bar.set_label(lang)
plt.legend(loc='upper right')
plt.xlabel('Sale')
plt.ylabel('Programming Languages')
plt.title('Sales of Books')
plt.show()
```

# 4b

```python
categories = ['Category1', 'Category2', 'Category3', 'Category4', 'Category5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
x = np.arange(len(categories))

plt.bar(x, means_men, label= 'Men')
plt.bar(x, means_women, bottom=means_men, label='Women')
plt.xlabel('Categories')
plt.ylabel('Means')
plt.title('Means by Category')
plt.xticks(x, categories)
for i in range(len(categories)):
    plt.text(x[i], means_men[i]/2, str(means_men[i]), ha='center', color='white')
    plt.text(x[i], means_men[i] + means_women[i]/2, str(means_women[i]), ha='center', color='black')
plt.legend()
plt.show()
```

# 5b

```python
data = pd.read_csv("company-sales.csv")

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data['total_profit'], bins=10, edgecolor='black')
plt.title('Histogram of Total Profit by Month')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
for count, bin_start in zip(n, bins):
    print(f"Range {bin_start}-{bin_start + (bins[1] - bins[0])}: {int(count)} occurences")
most_common_range = bins[n.argmax()]
print(f'\nThe most common profit range is {most_common_range}-{most_common_range + (bins[1] - bins[0])} with {int(n.max())} occurences')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

df = pd.read_csv("planets.csv")

plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(
    x='year', y='distance', hue='mass', style='number', size='mass',
    sizes=(20, 200), palette='viridis', data=df
)
scatter.set_title('Year-wise Distance Covered with Mass as Additional Features')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# 7a

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("planets.csv")

plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(
    x='year', y='distance', hue='mass', style='method', size='mass',
    sizes=(20, 200), palette='coolwarm', data=df
)
scatter.set_title('Year-wise Distance Covered with Mass and Method as Additional Features')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# 7b

```python
method_distance = df.groupby('method')['distance'].mean().reset_index()

plt.figure(figsize=(12, 6))
bar_chart = sns.barplot(x='method', y='distance', hue='method', dodge=False,
                        data=method_distance, palette='coolwarm', legend=False)
bar_chart.set_title('Average Distance Covered by Each Method')
bar_chart.set_xticks(range(len(method_distance['method'])))
bar_chart.set_xticklabels(method_distance['method'], rotation=45, horizontalalignment='right')
plt.show()
```
