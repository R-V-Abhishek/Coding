```python
'''Write program to Print all the nodes reachable from a given starting node in
a digraph using BFS method'''
```

```python
def BFS(adj_mat,start,n):
    visited=[False]*n
    queue=[]
    visited[start]=True
    queue.append(start)
    print(f"Nodes reachable from node{start}:")
    print(start,end=" ")
    while queue:
        curr=queue.pop(0)
        for i in range(n):
            if adj_mat[curr][i]==1 and not visited[i]:
                visited[i]=True
                queue.append(i)
                print(i,end=" ")
    print()
    
n=int(input("Enter the number of nodes:"))
adj_mat=[]
print("Enter the adjacency matrix:")
for i in range(n):
    row=list(map(int,input().split()))
    adj_mat.append(row)
start=int(input("Enter the start node:"))
if 0<=start<n:
    BFS(adj_mat,start,n)
else:
    print("Invalid Input")
    
```

```python
'''Write a program to check whether a given graph is connected or not using
DFS method.'''
```

```python
def DFS(adj_mat, start, n):
    visited = [False] * n
    stack = []
    stack.append(start)
    visited[start] = True
    
    while stack:
        curr = stack.pop()
        for i in range(n):
            if adj_mat[curr][i] == 1 and not visited[i]:
                visited[i] = True
                stack.append(i)
    
    # Check if all vertices are visited
    for i in range(n):
        if not visited[i]:  # Fixed: visited(i) -> visited[i]
            return False
    return True  # Fixed: moved outside the loop

def main():
    n = int(input("Enter the number of vertices of the graph: "))
    adj_mat = []
    print("Enter the adjacency matrix (row by row):")
    for i in range(n):
        row = list(map(int, input().split()))
        adj_mat.append(row)
    
    start = int(input("Enter the start vertex: "))
    if 0 <= start < n:  # Added validation
        if DFS(adj_mat, start, n):
            print("The graph is connected")
        else:
            print("The graph is not connected")
    else:
        print("Invalid start vertex")

if __name__ == "__main__":
    main()
```

```python
'''Sort a given set of N integer elements using Merge Sort technique.'''
```

```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left=arr[:mid]
    right=arr[mid:]
    
    left=merge_sort(left)
    right=merge_sort(right)
    
    return merge(left,right)
def merge(left,right):
    result=[]
    i=j=0
    while i<len(left) and j<len(right):
        if left[i]>=right[j]:
            result.append(right[j])
            j+=1
        else:
            result.append(left[i])
            i+=1
    while i<len(left):
        result.append(left[i])
        i+=1
    while j<len(right):
        result.append(right[j])
        j+=1
    return result
n=int(input("Enter the number of elements in the array:"))
print("Enter the numbers of array:")
arr=[]
for i in range(n):
    arr.append(int(input()))
result=merge_sort(arr)
for i in range(n):
    print(result[i],end=" ")
    
```

```python
'''Sort a given set of N integer elements using Quick Sort technique.'''
```

```python
def quick_sort(arr,low,high):
    if low<high:
        pivot=partition(arr,low,high)
        quick_sort(arr,low,pivot-1)
        quick_sort(arr,pivot+1,high)
def partition(arr,low,high):
    pivot=arr[low]
    i=low+1
    for j in range(low+1,high+1):
        if arr[j]<pivot:
            arr[i],arr[j]=arr[j],arr[i]
            i += 1
    arr[low],arr[i-1]=arr[i-1],arr[low]
    return i-1
n=int(input("Enter the number of elements in the array:"))
arr=[]
for i in range(n):
    arr.append(int(input()))
quick_sort(arr,0,n-1)
print("The sorted array is:")
for i in range(n):
    print(arr[i],end=" ")
```

```python
def heapify(arr,n,i):
    largest=i
    left_child=2*i+1
    right_child=2*i+2
    
    if left_child<n and arr[largest]<arr[left_child]:
        largest=left_child
    if right_child<n and arr[largest]<arr[right_child]:
        largest=right_child
    if largest!=i:
        arr[largest],arr[i]=arr[i],arr[largest]
        heapify(arr,n,largest)
def heapsort(arr):
    n=len(arr)
    for i in range(n//2-1,-1,-1):
        heapify(arr,n,i)
    for i in range(n-1,0,-1):
        arr[0],arr[i]=arr[i],arr[0]
        heapify(arr,i,0)
    
n=int(input("Enter the elements of the list: "))
arr=[]
for i in range(n):
    arr.append(int(input()))
heapsort(arr)
for i in arr:
    print(i,end=" ")
```

```python
'''Implement 0/1 Knapsack problem using dynamic programming.'''
```

```python
def knapsack(value, weight, capacity):
    n = len(weight)
    # Initialize the dp table properly first
    dp = [[0 for _ in range(capacity+1)] for _ in range(n+1)]
    
    # Now fill the dp table
    for i in range(1, n+1):
        for w in range(capacity+1):
            if weight[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], dp[i-1][w-weight[i-1]]+value[i-1])
            else:
                dp[i][w] = dp[i-1][w]
                
    # Rest of the code for tracking selected items
    selected_items = []
    i, j = n, capacity
    while i > 0 and j > 0:
        if dp[i][j] != dp[i-1][j]:
            selected_items.append(i-1)
            j -= weight[i-1]
        i -= 1
    selected_items.reverse()
    
    return selected_items, dp[n][capacity]
values=[]
weight=[]
n=int(input("Enter the number of items: "))
print("Enter the value of each element:")
for i in range(n):
    values.append(int(input()))
print("Enter the weights of the items:")
for i in range(n):
    weight.append(int(input()))
capacity=int(input("Enter the total capacity of the knapsack: "))
selected,max_value=knapsack(values,weight,capacity)
print(f"The max profit is:{max_value}")
print(f"The item selected are{selected}")
for i in selected:
    print(f"The item{i}:weight={weight[i]},value={values[i]}")
```

```python
def floyd_warshall(graph, n):
    # Create a distance matrix and initialize it with the input graph
    dist = [row[:] for row in graph]  # Deep copy of each row

    # Initialize the path matrix for path reconstruction
    path = [[-1 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if i != j and graph[i][j] != 999:
                path[i][j] = i

    # Main Floyd-Warshall algorithm
    for k in range(n):  # Consider each vertex as an intermediate
        for i in range(n):  # Consider each vertex as a start
            for j in range(n):  # Consider each vertex as an end
                # If going through vertex k provides a shorter path
                if dist[i][k] != 999 and dist[k][j] != 999:
                        dist[i][j] = min(dist[i][k] + dist[k][j], dist[i][j])
                        path[i][j] = path[k][j]

    return dist, path

def main():
    n = int(input("Enter the number of vertices: "))
    
    # Initialize the graph
    graph = []
    print("Enter the adjacency matrix (use weights for edges, 999 for no direct edge, 0 for self-loops):")
    for i in range(n):
        row = list(map(int, input().split()))
        graph.append(row)
    
    # Call Floyd-Warshall algorithm
    dist, path = floyd_warshall(graph, n)
    
    print("\nAll Pairs Shortest Paths:")
    for i in range(n):
        for j in range(n):
            if dist[i][j] == 999:
                print("INF", end="\t")
            else:
                print(dist[i][j], end="\t")
        print()
    
if __name__ == "__main__":
    main()
```

```python
def prim_mst(graph, n):
    # Initialize arrays for MST construction
    key = [999] * n  # Key values used to pick minimum weight edge
    parent = [-1] * n  # Array to store constructed MST
    mst_set = [False] * n  # To track vertices included in MST
    
    # Always include first vertex in MST
    key[0] = 0
    
    # The MST will have n vertices
    for count in range(n):
        # Pick the minimum key vertex from the set of vertices not yet included in MST
        min_key = 999
        min_index = -1
        for v in range(n):
            if not mst_set[v] and key[v] < min_key:
                min_key = key[v]
                min_index = v
        
        # Add the picked vertex to the MST set
        u = min_index
        mst_set[u] = True
        
        # Update key value and parent index of adjacent vertices
        for v in range(n):
            # Update only if edge exists (weight != 999), 
            # vertex v is not in MST, and weight is smaller than current key value
            if graph[u][v] != 999 and not mst_set[v] and graph[u][v] < key[v]:
                key[v] = graph[u][v]
                parent[v] = u
    
    return parent, key

def main():
    n = int(input("Enter the number of vertices: "))
    
    # Initialize the graph
    graph = []
    print("Enter the adjacency matrix (use weights for edges, 999 for no direct edge, 0 for self-loops):")
    for i in range(n):
        row = list(map(int, input().split()))
        graph.append(row)
    
    # Call Prim's algorithm
    parent, key = prim_mst(graph, n)
    
    # Calculate total cost
    total_cost = sum(key[i] for i in range(1, n))
    
    # Print the MST edges and their weights
    print("\nMinimum Spanning Tree Edges:")
    print("Edge \tWeight")
    for i in range(1, n):
        print(f"{parent[i]} - {i}\t{graph[parent[i]][i]}")
    
    print(f"\nTotal Cost of MST: {total_cost}")

if __name__ == "__main__":
    main()
```

```python
def dijkstra(graph, n, source):
    dist = [999] * n
    dist[source] = 0
    visited = [False] * n
    pred = [-1] * n

    for count in range(n):
        # Find the unvisited vertex with the smallest distance
        min_dist = 999
        u = -1
        for v in range(n):
            if not visited[v] and dist[v] < min_dist:
                min_dist = dist[v]
                u = v

        if u == -1:
            break

        visited[u] = True

        for v in range(n):
            if (not visited[v] and graph[u][v] != 999 and
                dist[u] + graph[u][v] < dist[v]):
                dist[v] = dist[u] + graph[u][v]
                pred[v] = u

    return dist, pred

def print_path(pred, dest, path=None):
    if path is None:
        path = []
    if dest == -1:
        return path
    print_path(pred, pred[dest], path)
    path.append(dest)
    return path

def main():
    n = int(input("Enter the number of vertices: "))
    
    # Initialize the graph
    graph = []
    print("Enter the adjacency matrix (use weights for edges, 999 for no direct edge, 0 for self-loops):")
    for i in range(n):
        row = list(map(int, input().split()))
        graph.append(row)
    
    source = int(input("Enter the source vertex: "))
    
    # Validate the source vertex
    if source < 0 or source >= n:
        print("Invalid source vertex")
        return
    
    # Call Dijkstra's algorithm
    dist, pred = dijkstra(graph, n, source)
    
    # Print the shortest distances and paths
    print("\nShortest distances from vertex", source, ":")
    for i in range(n):
        if dist[i] == 999:
            print(f"Vertex {i}: No path exists")
        else:
            path = print_path(pred, i)
            print(f"Vertex {i}: Distance = {dist[i]}, Path = {' â†’ '.join(map(str, path))}")

if __name__ == "__main__":
    main()
```

```python
'''3 Write program to obtain the Topological ordering of vertices in a given digraph.'''
```

```python
def topological_sort(adj_mat, n):
    in_degree = [0] * n
    for i in range(n):
        for j in range(n):
            if adj_mat[i][j]:
                in_degree[j] += 1

    queue = []
    for i in range(n):
        if in_degree[i] == 0:
            queue.append(i)

    topo_order = []
    while queue:
        u = queue.pop(0)
        topo_order.append(u)
        for v in range(n):
            if adj_mat[u][v]:
                in_degree[v] -= 1
                if in_degree[v] == 0:
                    queue.append(v)

    if len(topo_order) != n:
        print("The graph has a cycle. Topological ordering is not possible.")
    else:
        print("Topological Ordering of the vertices:")
        print(" ".join(map(str, topo_order)))

n = int(input("Enter the number of vertices: "))
adj_mat = []
print("Enter the adjacency matrix (row by row):")
for i in range(n):
    row = list(map(int, input().split()))
    adj_mat.append(row)

topological_sort(adj_mat, n)
```
